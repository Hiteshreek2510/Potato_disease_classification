```python
import pdfplumber as plum
import numpy as np
import pandas as pd
```

```python
data=[]
with plum.open("boisar.pdf") as pdf:
    for page in pdf.pages:
        tables=page.extract_table()
        if tables :
            data.extend(tables)
data_array=np.array(data,dtype=object)
df=pd.DataFrame(data_array)
df.to_csv('out1.csv',index=False,header=True)
        
```
